## Openning Data

```julia
using MAT
using LinearAlgebra
using SparseArrays
using JuMP
using GLPK
```

```julia
file = matopen("R1T2.mat")
varnames = names(file)
```

```julia
S = read(file, "S")
l1 = read(file, "l1")
u1 = read(file, "u1")
```

```julia
l1_nzval = l1["nzval"]
l1_nzind = l1["nzind"]
l1_n = l1["n"]
l = SparseVector(l1_n, l1_nzind, l1_nzval)
```

```julia
u1_nzval = u1["nzval"]
u1_nzind = u1["nzind"]
u1_n = u1["n"]
u = SparseVector(u1_n, u1_nzind, u1_nzval)
```

```julia
m,n = size(S)
```

```julia
mzeros = zeros(m,1);
```

# Constructing  Model

```julia
model = Model(GLPK.Optimizer)
@variable(model,v[1:n])
@constraint(model, l .<= v)
@constraint(model, v .<= u)
@constraint(model, S*v .== mzeros)
@objective(model,Min,1)
```

```julia
optimize!(model)
```

```julia
@show termination_status(model)
@show primal_status(model)
@show dual_status(model)
@show objective_value(model)
@show JuMP.value.(v)
```
